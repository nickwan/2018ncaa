```python
def get_team_data(TEAM, OPP_TEAM):
    temp = pd.DataFrame()
    game_denom_tm = season.loc[season['Team2']==TEAM, 'o_G'].values
    temp['o_SRS'] = [(season.loc[season['Team2']==TEAM, 'o_SRS']).values[0]]
    temp['Tm_FGA'] = (season.loc[season['Team2']==TEAM, 'totals_FGA'] / game_denom_tm).values[0]
    temp['Tm_3PA'] = (season.loc[season['Team2']==TEAM, 'totals_3PA'] / game_denom_tm).values[0]
    temp['Tm_FTA'] = (season.loc[season['Team2']==TEAM, 'totals_FTA'] / game_denom_tm).values[0]
    temp['Tm_ORB'] = (season.loc[season['Team2']==TEAM, 'totals_ORB'] / game_denom_tm).values[0]
    temp['Tm_TRB'] = (season.loc[season['Team2']==TEAM, 'totals_TRB'] / game_denom_tm).values[0]
    temp['Tm_STL'] = (season.loc[season['Team2']==TEAM, 'totals_STL'] / game_denom_tm).values[0]
    temp['Tm_BLK'] = (season.loc[season['Team2']==TEAM, 'totals_BLK'] / game_denom_tm).values[0]
    temp['Tm_TOV'] = (season.loc[season['Team2']==TEAM, 'totals_TOV'] / game_denom_tm).values[0]
    temp['Tm_PF'] = (season.loc[season['Team2']==TEAM, 'totals_PF'] / game_denom_tm).values[0]

    game_denom_opp = season.loc[season['Team2']==OPP_TEAM, 'o_G'].values
    temp['o_SRS_opp'] = (season.loc[season['Team2']==OPP_TEAM, 'o_SRS']).values[0]
    temp['opp_FGA'] = (season.loc[season['Team2']==OPP_TEAM, 'totals_FGA'] / game_denom_opp).values[0]
    temp['opp_3PA'] = (season.loc[season['Team2']==OPP_TEAM, 'totals_3PA'] / game_denom_opp).values[0]
    temp['opp_FTA'] = (season.loc[season['Team2']==OPP_TEAM, 'totals_FTA'] / game_denom_opp).values[0]
    temp['opp_ORB'] = (season.loc[season['Team2']==OPP_TEAM, 'totals_ORB'] / game_denom_opp).values[0]
    temp['opp_TRB'] = (season.loc[season['Team2']==OPP_TEAM, 'totals_TRB'] / game_denom_opp).values[0]
    temp['opp_STL'] = (season.loc[season['Team2']==OPP_TEAM, 'totals_STL'] / game_denom_opp).values[0]
    temp['opp_BLK'] = (season.loc[season['Team2']==OPP_TEAM, 'totals_BLK'] / game_denom_opp).values[0]
    temp['opp_TOV'] = (season.loc[season['Team2']==OPP_TEAM, 'totals_TOV'] / game_denom_opp).values[0]
    temp['opp_PF'] = (season.loc[season['Team2']==OPP_TEAM, 'totals_PF'] / game_denom_opp).values[0]
    return temp
```
